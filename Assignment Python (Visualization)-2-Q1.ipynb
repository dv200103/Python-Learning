```python
'''

Load the titanic_original data file

 (i) Create a pie chart presenting the male/female proportion 

 (ii) Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

'''
```

```python
import pandas as pd
from matplotlib import pyplot as plt
titanic_original = pd.read_csv("titanic_original.csv")
titanic_original
```

```python
# (i)
titanic_original["sex"].value_counts().plot(kind = "pie", autopct = "%1.1f%%")
plt.show()
```

```python
# (ii)
import seaborn as sns
sns.scatterplot(data = titanic_original, x = "fare", y = "age", hue = "sex")
plt.show()
```
