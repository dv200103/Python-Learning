```python
'''
Load the necessary package for plotting using pyplot from matplotlib. Example - Days(x-axis) represents 8 days and Speed represents a carâ€™s speed.
Plot a Basic line plot between days and car speed, put x axis label as days and y axis label as car speed and put title Car Speed Measurement.

         Days=[1,2,3,4,5,6,7,8]

         Speed=[60,62,61,58,56,57,46,63]
'''
```

```python
# import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings("ignore")
Days=[1,2,3,4,5,6,7,8]
Speed=[60,62,61,58,56,57,46,63]
plt.xlabel("days")
plt.ylabel("car speed")
plt.title("Car Speed Measurement")
plt.plot(Days,Speed)
plt.show()
```
