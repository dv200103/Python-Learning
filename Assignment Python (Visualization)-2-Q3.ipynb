```python
'''

The Planets dataset contains details about the 1,000+ extrasolar planets discovered up to 2014. 
Visualize the distribution of the masses of the planets (expressed as a multiple of the mass of Jupiter), using a histogram and a box plot. 
Make appropriate modifications to the chart title, axis titles, legend, figure size, font size, colors etc. 
to make the chart readable and visually appealing.

'''
```

```python
import pandas as pd
df = pd.read_csv("planets.csv")
df
```

# Histogram

```python
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings("ignore")
plt.figure(figsize = (9,9))
plt.hist(df["mass"],bins = 10, color = "orange")
plt.xlabel("Mass", fontsize = 14, weight = "bold", color = "green" )
plt.ylabel("Frequency of Mass", fontsize = 14, weight = "bold", color = "green" )
plt.title("Frequency distribution of Mass", fontsize = 14, weight = "bold", color = "green" )
plt.show()
```

# Boxplot

```python
import seaborn as sns
sns.boxplot(data = df,  x = "mass")
plt.xlabel("Mass", fontsize = 14, color = "green", weight = "bold")
plt.title("Distribution of planet masses", fontsize = 14, color = "Red", weight = "bold")
plt.show()
```
