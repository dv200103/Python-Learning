## Multiline plot

```python
x = [10,25,35,45,65,85]
y1 = np.sqrt(x)
y2 = np.log(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y1,marker="v")
plt.plot(x,y2,marker="v")
plt.xlabel("The values of x")
plt.ylabel("The values of y")
plt.title("The distribution of log and sqrt of x")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y1,marker="v",color='red')
plt.plot(x,y2,marker="v",color='green')
plt.xlabel("The values of x")
plt.ylabel("The values of y")
plt.title("The distribution of log and sqrt of x")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y1,marker="v",color='red',label='sqrt')
plt.plot(x,y2,marker="v",color='green',label='log')
plt.xlabel("The values of x")
plt.ylabel("The values of y")
plt.title("The distribution of log and sqrt of x")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y1,marker="v",color='red',label='sqrt')
plt.plot(x,y2,marker="v",color='green',label='log')
plt.legend()
plt.xlabel("The values of x")
plt.ylabel("The values of y")
plt.title("The distribution of log and sqrt of x")
plt.grid()
plt.show()
```
