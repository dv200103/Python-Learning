# DataFrame indexing and slicing

```python
# Whenever we need to deal with column in DataFrame then use iloc or go by normal way.

# Whenever we need to deal with row and column by indexes in DataFrame then use iloc.

# Whenever we need to deal with row in DataFrame then use loc.

# When we have to deal with both row and column by indexes and key or column name then use loc.
```

```python
data = {"Name":["Google","Facebook","IBM","Microsoft"],"Date":["12-Dec-2018","12-Dec-2018","12-Dec-2018","12-Dec-2018"],"Shares":["70%","15%","2%","13%"],"Rank":[1,2,3,4]}
df = pd.DataFrame(data)
df
```

```python
df["Salary"] = [100000,120000,130000,150000] # Adding a column to a existing DataFrame.
df
```

```python
df.loc[4] = ["Amazon","12-Dec-2019","22%",5,90000] # Adding new row in DataFrame.
df
```

```python
df[["Name"]] # Slicing the DataFrame for a single column by keywise or columnwise from DataFrame.
```

```python
df[["Name","Rank"]] # Slicing the DataFrame for a multiple column by keywise or columnwise from DataFrame.
```

```python
df.iloc[2] # Accessing row information df.iloc[index_value].
```

```python
df.iloc[0:2] # Slicing the DataFrame only by row indexing from DataFrame.(If column indexing not mention then it will take all column as a by default.)
```

```python
df.iloc[:,0:2] # Slicing the DataFrame only by column indexing from DataFrame.(If row indexing not mention then it will take all row as a by default.)
```

```python
df.iloc[0:2,0:2] # Slicing the DataFrame by row and column indexing from DataFrame.
```

```python
df.iloc[2,3] # Accessing a particular cell value df.iloc[row_index,column_index]
```

```python
df.iloc[3,4] # Slicing a specific row and a specific column(index wise of both row and column) or slicing a specific value from DataFrame.
```

```python
df.loc[2] # Slicing the specific row for all the column from DataFrame.
```

```python
df.loc[1:3] # Slicing the multiple rows between a range for all the columns from DataFrame.
```

```python
df.loc[3,"Salary"] # Slicing a specific row and a specific column(keywise or column name) or slicing a specific value from DataFrame.
```

```python
df.loc[0:1,"Name"] # Slicing the multiple rows between a range for specific column from DataFrame.
```

```python
df.loc[0:3,"Name"] # Slicing the DataFrame by row indexing and keywise or columnwise from DataFrame.
```

# Pandas Function

# Arithmetic operations

```python
import pandas as pd 
dff = pd.DataFrame({"A":[1,2,3],"B":[4,5,6]})

# Addition
dff["C"] = dff["A"] + dff["B"]

# Subtraction
dff["D"] = dff["A"] - dff["B"]

# Multiplication
dff["E"] = dff["A"] * dff["B"]

# Division
dff["F"] = dff["A"] / dff["B"]

dff
```

# Aggregation Functions

```python
total = dff["A"].sum() # Pandas provides aggregation functions like sum(), mean(), median(), min(), and max() that can be applied to columns of a DataFrame or Series.
total
```

```python
avg = dff["B"].mean()
avg
```

```python
median = dff["C"].median()
median
```

```python
min_val = dff["D"].min()
min_val
```

```python
max_val = dff["E"].max()
max_val
```

# Dealing With NaN values

```python
import pandas as pd
Movie = pd.read_csv("Movies1.csv")
Movie
```

```python
Movie["worlwide_gross_income"].isnull() # Returns True where null values are present.
```

```python
Movie["worlwide_gross_income"].isnull().head() # Check Null or NaN values for only top 5 rows due to head() and Return True.
```

```python
Movie["worlwide_gross_income"].notnull() # Returns True where null values are not present, it works opposite of isnull()
```

```python
Movie["worlwide_gross_income"].notnull().head() # Returns True where null values are not present for only top 5 rows due to head() 
```

```python
Movie[Movie["worlwide_gross_income"].isnull()] # To display the rows with null values in worlwide_gross_income column.
```

```python
Movie[Movie["worlwide_gross_income"].isnull()].head() # To display the rows with null values in worlwide_gross_income column for only top 5 rows due to head(). 
```

```python
Movie[Movie["worlwide_gross_income"].isnull()].tail() # To display the rows with null values in worlwide_gross_income column for only bottom 5 rows due to tail().
```

```python
Movie[Movie["worlwide_gross_income"].notnull()] # To display the rows which doesn't contain null values in worlwide_gross_income column.
```

```python
Movie[Movie["worlwide_gross_income"].isnull()].fillna("NA")
```

```python
Movie[Movie["worlwide_gross_income"].isnull()]
```

```python
Movie.dropna() # Drop the rows where Null values is present in DataFrame, attribute how = "any" drop the all the rows where Null values is present , By default it is how = "any", and how = "all" drop the row if that row completely filled with Null values.
```

```python
Movie
```

```python
Movie.dropna(how = "all") # No row contain only null values that's why output doesn't change and zero rows get effected or drop.
# We can give axis = 0 in dropna for column drop according to null values, if column contain null values then it will drop that column, by default axis = 0 that's why it always drop row according to null values.
```

```python
Movie.dropna(how = "any", axis = 1) # Droping column according to Null values.
```
