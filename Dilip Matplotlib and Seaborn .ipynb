# Subplot: Plots within a plot

```python
from matplotlib import pyplot as plt # Easy way
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
plt.subplot(2,1,1) # 2 represents 2 horizontal row and 1 represents 1 vertical column and 1 represents 1st place.
plt.plot(x,y, color = "blue", label = "x-y plot")
plt.legend()
plt.subplot(2,1,2) # 2 represents 2 horizontal row and 1 represents 1 vertical column and 2 represents 2nd place.
plt.plot(x,z, color = "red", label = "x-z plot")
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt # Easy way
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
plt.subplot(2,1,2) # 2 represents 2 horizontal row and 1 represents 1 vertical column and 2 represents 2nd place.
plt.plot(x,y, color = "blue", label = "x-y plot")
plt.legend()
plt.subplot(2,1,1) # 2 represents 2 horizontal row and 1 represents 1 vertical column and 1 represents 1st place.
plt.plot(x,z, color = "red", label = "x-z plot")
plt.legend()
plt.show() # Opposite output compare to above cell due to changes in the third place value in plt.plot(2,1,1) and plt.plot(2,1,2)
```

# BOXPLOT Calculation

```python
import seaborn as sns
import numpy as np
dataset = (4,6,75,22,46,3,6,8,2,222,44,23,56,-2,-9,1111)
dataset 
sorted_data = sorted(dataset)
sorted_data
```

```python
np.median(sorted_data) # Median(Q2)
```

```python
perc1, perc2 = np.percentile(sorted_data,[25,75]) # Lower Quartile(Q1) and upper Quartile(Q3)
```

```python
perc1 # Lower Quartile(Q1)
```

```python
perc2 # upper Quartile(Q3)
```

```python
iqr = perc2 - perc1 # Inter Quartile Range[IQR]
iqr
```

```python
lower = perc1 - (1.5*iqr) # Lower limit (lower Whisker)
lower
```

```python
upper = perc2 + (1.5*iqr) # upper limit (upper Whisker)
upper
```

```python
lower_val = []
upper_val = []
for i in sorted_data:
    if i < lower:
        lower_val.append(i)
    elif i > upper:
        upper_val.append(i)
        
print(f"Lower Values: {lower_val}")
print(f"Upper Values: {upper_val}") # Extreme low outliers and extreme high outliers.
```
