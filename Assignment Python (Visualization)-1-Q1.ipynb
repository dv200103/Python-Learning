```python
from matplotlib import pyplot as plt
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
plt.subplot(2,1,1)
plt.plot(x,y, color = "blue", label = "x-y plot")
plt.legend()
plt.subplot(2,1,2)
plt.plot(x,z, color = "red", label = "x-z plot")
plt.legend()
plt.show()
```
