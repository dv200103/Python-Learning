```python
from matplotlib import pyplot as plt
import numpy as np
fig, (ax1,ax2) = plt.subplots(2,1, figsize=(8,8))
x = np.arange(0,100)
y = x*2
z = x**2
ax1.plot(x,y, color = "Red")
ax2.plot(x,z, color = "Brown")
plt.show()
```
