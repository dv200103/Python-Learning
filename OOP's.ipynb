```python
# Multilevel inheritance
class Hospital:
    def __init__(self, patient_name, doctor_name): # Default Constructor
        self.p_name = patient_name
        self.d_name = doctor_name
        print("Default constructor has run")

        
class OPD(Hospital): # Child Class , accessing the Parent class -Hospital 
    def __init__(self, patient_name, doctor_name, Gynecology, General_Medicine): 
        super().__init__(patient_name, doctor_name)
        Hospital.__init__(self, patient_name, doctor_name)
        self.opd_Gyn = Gynecology
        self.opd_Ger = General_Medicine
        print("CHILD cONSTRUCTOR HAS RUN")
        
    def opd(self):
        print(f'Patient {self.p_name} is charged {self.opd_Gyn} by {self.d_name} for the treatment of {self.opd_Ger}')
        
    def display(self):
        print(f"{self.p_name}")


Patient1 = Hospital('Kshitij', 'Sai')
Patient2=OPD("Verma","Khan","G1","Paracetamol")
Patient2.opd()
```
